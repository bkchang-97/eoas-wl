# How to md2canvas

By: Benjamin Chang

- md2canvas is a Python script that allows for a markdown file to be directly exported to Canvas as a Quiz which bypasses the instructor's need to manually set one up 
- questions can be of various formats: MC, short answer, etc. each with its own specific structure and syntax
- various scripts and tools have been created in support of md2canvas creating a workflow

1. Depending on the file format you are working with, Phil has written multiple scripts that can covert each of them into a markdown file: `read_pptx.py`, `run_pandoc.py`, etc located in the `scripts` folder in e340quizzes
2. Following that, process the markdown file one more time through the `write_meta.py` script or `write_clickers.py` depending on if you're interesting in creating a question out of each page/slide or only extracting clicker questions
    
    *python write_meta.py "starting file" "name of output file"*

3. Now that you have quiz questions in a markdown file, you can start editing the markdown file in order to specify what kind of questions, correct answers, images/figures, etc. The samplequiz.md file in md2canvas have great examples of how to create different questions and how to format them. In the case of pptx slides, write_meta will format each slide as a question but write_clickers filters for clicker question specific slides.
